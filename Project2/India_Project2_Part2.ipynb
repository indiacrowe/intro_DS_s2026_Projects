## Data Wrangling
Adding a new column based on another

```python
hi_new=human_interaction.copy()
hi_new
```

```python
def drained(Time_spent_Alone):
    if pd.isnull(Time_spent_Alone):
        return 'N/A'
    elif Time_spent_Alone < 5:
        return 1
    else:
        return 2

def fright(Friends_circle_size):
    if pd.isnull(Friends_circle_size):
        return 'N/A'
    elif Friends_circle_size <= 5:
        return 2
    else:
        return 1
```

```python
hi_new['Drained_after_events']=hi_new['Time_spent_Alone'].apply(drained)
hi_new['Stage_fright']=hi_new['Friends_circle_size'].apply(fright)
hi_new
```

```python
hi_new.head(11)
```

The new column Drained after events is based on the person's rating of how much time they spend alone. If they spend less than 5 hours
alone, they're considered extroverts who rarely feel drained after socializing. So those who are considered extroverts are given the
number 1, in case a correlation table is needed later on. And if they have 5 or less friends they are consdered introverts, so they are given
the number 2.

## Summary and Conclusion

One interesting aspect of this dataset is the almost textook contrast between how
introverts and extroverts manage their social energy and time. A clear pattern
appeared during the initial exploration: introverts report spending over three times as
much time alone as extroverts, while extroverts maintain significantly larger friend
circles and attend more social events. However, a few questions arise regarding the
data's distribution, particularly, the Friends_circle_size column shows some people
with zero friends maintain a high Post_frequency on social media, which suggests
that digial social engagement doesn't always mirror physical social circles.
Additionally, the lack of 'neutral' or 'Ambivert' categories rises the question of
whether the data was intentionally simplfied for classification puroses or if individuals
were forced to choose between those two options.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,6))
corr_matrix=hi_new.corr(numeric_only=True)
ax=sns.heatmap(corr_matrix, annot=True, fmt='.3f', cmap='GnBu', center=0, 
linewidths=0.5, cbar_kws={'label':'Correlatation'})
plt.title('Correlation Heat Map of Numeric Features')
plt.savefig('P2/Heatmap.jpg')
plt.show()
```

This heatmap shows how different social behaviors are related to each other. The main pattern is that time spent alone is strongly negatively related to things like going outside, social event attendance, and friend group size, meaning the more time someone spends alone, the less socially active they tend to be. On the other hand, those social activities are all positively related to each other, so people who go out more also tend to have more friends and attend more events. It also shows that feeling drained and having stage fright are positively related to spending more time alone and less social activity.
